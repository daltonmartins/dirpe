```python
# Importando a BeautifulSoup e Urllib para requisição dos dados
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
import re

#tabela de dados de saída com os parâmetros a serem analisados
tabela = pd.DataFrame()

# carrega a lista de grupos de pesquisa
listagrupos = pd.read_excel('lista136ultimos.xlsx', index_col=None, header=None) #o parâmetro index_col e header servem para planilha sem título de coluna e cabeçalho

quantidadegrupos = len(listagrupos[0])

for grupo in listagrupos[0]:

    # lendo a URL com a urllopen
    html = urlopen(grupo)

    # Enfim mostrando o poder da bs4
    bs = BeautifulSoup(html, 'lxml')

    tags = bs.find_all('div', {'class': ['controls']}) #as DIVs que possuem os valores de parâmetros são as CONTROLS. O resto são tabelas que serão coletadas com Pandas

    contador=0 # serve para iterar entre todas as DIVs encontradas no comando anterior
    for tag in tags:
        if (contador==1):
            anoformacao=tag.get_text().strip()
        elif (contador==4):
            lideres=tag.get_text()
            lideres=lideres.replace("Permite enviar email$(function(){PrimeFaces.cw('Tooltip','widget_idFormVisualizarGrupoPesquisa_j_idt121',{id:'idFormVisualizarGrupoPesquisa:j_idt121',showEffect:'fade',hideEffect:'fade',target:'idFormVisualizarGrupoPesquisa:idCommandButtonEmail'});});", '')
            lideres=lideres.replace('\t','').split('\n')
            lideres.pop()
        elif (contador==5):
            areas=tag.get_text().strip()
        elif (contador==7):
            unidade=tag.get_text().strip()
        elif (contador==23):
            email=tag.get_text().strip()
        elif (contador==21):
            telefone=tag.get_text().strip()
        elif (contador==26):
            nomedogrupo=tag.get_text().strip()
        contador=contador+1
      
    #TODAS AS TABELAS PODEM SER RECUPERADAS USANDO O PANDAS - 
    #DE REPERCUSSAO DO GRUPO PARA FRENTE É TUDO TABELA NO DGP
    dfs = pd.read_html(str(grupo), header=0)
    
    if (bs.find_all('div', {'id': ['repercussao']})):
        totalestudantes=dfs[1]['Quantidade de Estudantes'].sum()
        totalpesquisadores=dfs[1]['Quantidade de Pesquisadores'].sum()
        linhasdepesquisa=dfs[1]['Nome da linha de pesquisa'].tolist() 
        listapesquisadores=dfs[2]['Pesquisadores'].tolist() 
        listaestudantes=dfs[3]['Estudantes'].tolist() 
        listacolaboradorestrangeiro=dfs[5]['Colaboradores estrangeiros'].tolist() 
        listapaisescolaboradorestrangeiro=dfs[5]['País'].tolist() 
        listadatacolaboradoresestrangeiro=dfs[5]['Data inclusão'].tolist() 
        instituicoesparceiras=dfs[9]['Nome da Instituição Parceira'].tolist()
        listaequipamentos=dfs[11]['Equipamentos'].tolist()
        listasoftwares=dfs[12]['Softwares'].tolist()
    else:
        totalestudantes=dfs[0]['Quantidade de Estudantes'].sum()
        totalpesquisadores=dfs[0]['Quantidade de Pesquisadores'].sum()
        linhasdepesquisa=dfs[0]['Nome da linha de pesquisa'].tolist() 
        listapesquisadores=dfs[1]['Pesquisadores'].tolist() 
        listaestudantes=dfs[2]['Estudantes'].tolist() 
        listacolaboradorestrangeiro=dfs[4]['Colaboradores estrangeiros'].tolist() 
        listapaisescolaboradorestrangeiro=dfs[4]['País'].tolist() 
        listadatacolaboradoresestrangeiro=dfs[4]['Data inclusão'].tolist() 
        instituicoesparceiras=dfs[8]['Nome da Instituição Parceira'].tolist()
        listaequipamentos=dfs[10]['Equipamentos'].tolist()
        listasoftwares=dfs[11]['Softwares'].tolist()
    
    #cria o dicionário com os resultados coletados
    obj = {'Nome do grupo':nomedogrupo , 'Link do DGP':grupo,'Ano formação':anoformacao, 'Líderes': lideres, 'Áreas':areas,'Unidade acadêmica':unidade,'Email':email,'Telefone':telefone,'Linhas de pesquisa': linhasdepesquisa,'Estudantes':listaestudantes,'Total de Estudantes':totalestudantes,'Pesquisadores': listapesquisadores,'Total de Pesquisadores':totalpesquisadores,'Colaboradores estrangeiro':listacolaboradorestrangeiro,'Países estrangeiros':listapaisescolaboradorestrangeiro,'Data de colaboradores estrangeiro':listadatacolaboradoresestrangeiro,'Softwares':listasoftwares,'Equipamentos':listaequipamentos,'Instituicoes parceiras':instituicoesparceiras}

    #incrementa uma linha no dataframe
    tabela=tabela.append(obj, ignore_index=True)

    # Informa situação da coleta na tela
    quantidadegrupos=quantidadegrupos-1
    print(grupo,"... concluído. Faltam...:",quantidadegrupos)

tabela.to_excel('coletagrupos1.xlsx')
```
